```python
import csv

with open('PLL_Player_Gamelogs_Raw.csv') as f:
    data = [{k: str(v) for k, v in row.items()}
        for row in csv.DictReader(f, skipinitialspace=True)]
```

```python
import pandas as pd

df = pd.read_csv("PLL_Player_Gamelogs_Raw.csv")
df2 = df[['Player','Goals','Shots','ShotPct','ShotsOnGoal','ShotsOnGoalPct','position','year','SeasonSegment']].copy()
df2 = df2[df2['position'].isin(['A','M'])]
df2 = df2[df2['SeasonSegment'].isin(['SeasonSegmentRegseason'])]
df_2019 = df2[df2['year'].isin([2019])]
df_2020 = df2[df2['year'].isin([2020])]
df_2021 = df2[df2['year'].isin([2021])]
```

```python
import tabulate 
by_player_2019 = df_2019.groupby('Player',as_index=False).agg({'Goals':'sum','Shots':'sum','ShotPct':'mean','ShotsOnGoal':'sum','ShotsOnGoalPct':'mean'})
#by_player_2019
by_player_2019 = by_player_2019.loc[by_player_2019['Shots'] > 0]
by_player_2019
```

```python
by_player_2020 = df_2020.groupby('Player',as_index=False).agg({'Goals':'sum','Shots':'sum','ShotPct':'mean','ShotsOnGoal':'sum','ShotsOnGoalPct':'mean'})
#by_player_2020
by_player_2020 = by_player_2020.loc[by_player_2020['Shots'] > 0]
by_player_2020
```

```python
by_player_2021 = df_2021.groupby('Player',as_index=False).agg({'Goals':'sum','Shots':'sum','ShotPct':'mean','ShotsOnGoal':'sum','ShotsOnGoalPct':'mean'})
#by_player_2021
by_player_2021 = by_player_2021.loc[by_player_2021['Shots'] > 0]
by_player_2021
```
